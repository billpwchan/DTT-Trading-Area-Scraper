```python
import json

input1 = json.load(open('Input1.json', 'r', encoding='UTF-8'))
input2 = json.load(open('Input2.json', 'r', encoding='UTF-8'))
input3 = json.load(open('Input3.json', 'r', encoding='UTF-8'))
input4 = json.load(open('Input4.json', 'r', encoding='UTF-8'))
input5 = json.load(open('Input5.json', 'r', encoding='UTF-8'))
input6 = json.load(open('Input6.json', 'r', encoding='UTF-8'))


print(input1[0])
print(input2[0])
print(input3[0])
print(input4[0])
print(input5[0])
print(input6[0])
```

```python
parsed_list = []
for i in range(len(input1)):
    item = {}
    item['district'] = input1[i]['district']
    item['type'] = input1[i]['type']
    item['lng'] = input3[i]['lng']
    item['lat'] = input3[i]['lat']
    item['trading_area'] = input1[i]['trading_area']
    item['school_num'] = input1[i]['GaodengyuanxiaoNum']
    item['similar_shops_num'] = input2[i]['SimilarShop']
    item['human_traffic'] = input3[i]['PeopleNumber']
    item['subway_num'] = input4[i]['SubwayNum']
    item['bus_station_num'] = input4[i]['BusStationNum']
    item['residence_num'] = input5[i]['ResidenceNum']
    item['office_building_num'] = input6[i]['OfficeBulidingNum']

    parsed_list.append(item)
```

```python
parsed_list[0]
```

```python
import pandas as pd

trading_area = pd.DataFrame(parsed_list)
trading_area[['lng', 'lat', 'school_num', 'similar_shops_num', 'human_traffic', 'subway_num', 'bus_station_num', 'residence_num', 'office_building_num']] = trading_area[['lng', 'lat', 'school_num', 'similar_shops_num', 'human_traffic', 'subway_num', 'bus_station_num', 'residence_num', 'office_building_num']].apply(pd.to_numeric)
trading_area.head()
```

```python
types = trading_area.type.unique().tolist()
```

```python
trading_area[trading_area['type'] == '4-优质商圈']
```

```python
trading_area['school_num_rank'] = trading_area['school_num'].rank(ascending=False)
trading_area['similar_shops_num_rank'] = trading_area['similar_shops_num'].rank(ascending=False)
trading_area['human_traffic_rank'] = trading_area['human_traffic'].rank(ascending=False)
trading_area['subway_num_rank'] = trading_area['subway_num'].rank(ascending=False)
trading_area['bus_station_num_rank'] = trading_area['bus_station_num'].rank(ascending=False)
trading_area['residence_num_rank'] = trading_area['residence_num'].rank(ascending=False)
trading_area['office_building_num_rank'] = trading_area['office_building_num'].rank(ascending=False)

trading_area.head()
```

```python
trading_area['total_score'] = trading_area['school_num_rank'] + trading_area['similar_shops_num_rank'] + trading_area['human_traffic_rank'] + trading_area['subway_num_rank'] + trading_area['bus_station_num_rank'] + trading_area['residence_num_rank'] + trading_area['office_building_num_rank']

trading_area[trading_area['type'] == '4-优质商圈'].sort_values('total_score').head(10)
```

```python
trading_area_cbd = trading_area[trading_area['type'] == '4-优质商圈']

for i in range(trading_area_cbd['human_traffic'].min(), trading_area_cbd['human_traffic'].max(), 1000):
    for j in range(trading_area_cbd['subway_num'].min(), trading_area_cbd['subway_num'].max()):
        if (trading_area_cbd[(trading_area_cbd['human_traffic'] > i) & (trading_area_cbd['subway_num'] > j)]['trading_area'].count() <= 10):
            print(f'Human Traffic: {i}, Subway Num: {j} ')
            print('Number of Record: %d', trading_area_cbd[(trading_area_cbd['human_traffic'] > i) & (trading_area_cbd['subway_num'] > j)]['trading_area'].count())
            print(trading_area_cbd[(trading_area_cbd['human_traffic'] > i) & (trading_area_cbd['subway_num'] > j)])
            break
```

```python
trading_area_cbd['district_human_traffic_rank'] = trading_area_cbd.groupby('district')['human_traffic'].rank(ascending=False)
trading_area_cbd = trading_area_cbd.sort_values('district_human_traffic_rank', ascending=False)
top2 = trading_area_cbd.groupby('district').head(2).sort_values('district')
top2[['district', 'trading_area']]
```
